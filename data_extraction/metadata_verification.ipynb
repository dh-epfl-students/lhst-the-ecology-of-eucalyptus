# Gallica date distribution

```python
tot_documents = []
for year in range(1780, 1920):
    #time.sleep(15)
    query = f'gallica all "et" and dc.date = "{year}" sortby dc.date/sort.descending'
    doc = search.single_search_gallica(query, startRecord=0)
    try:
        tot_documents.append(int(doc.find("srw:numberofrecords").text.strip()))
    except:
        print(year)
        print(doc)
        tot_documents.append(0)
    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame()
df["year"] = range(1780, 1920)
df["docs"] = tot_documents
sns.histplot(df, x=df["year"], weights=df["docs"], bins=70, kde=True)
```
