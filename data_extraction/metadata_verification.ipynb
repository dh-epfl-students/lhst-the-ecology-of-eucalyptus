```python
import pandas as pd
import numpy as np
from tabulate import tabulate
```

```python
df = pd.read_csv("../data/document_data_clean_filtered.csv", encoding="utf-8")
```

```python
#fix random seed
np.random.seed(1)
```

### date

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.max_colwidth', 110):
    print(tabulate(df.sample(100)[["date_clean", "date"]]))
```

### author name

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.max_colwidth', 110):
    print(tabulate(df.dropna(subset="author_name_clean").sample(100)[["author_name_clean", "author"]]))
```

### author_type

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.max_colwidth', 110):
    print(tabulate(df.dropna(subset="author_type_clean").sample(100)[["author_type_clean", "author"]]))
```

### author_date

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.max_colwidth', 110):
    print(tabulate(df.dropna(subset="author").sample(100)[["author_birth_clean", "author_death_clean", "author"]]))
```

### publisher_name

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.max_colwidth', 110):
    print(tabulate(df.dropna(subset="publisher").sample(100)[["publisher_name_clean", "publisher"]]))
```

### publisher_place

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.max_colwidth', 110):
    print(tabulate(df.dropna(subset="publisher").sample(100)[["publisher_place_clean", "publisher"]]))
```

### type

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None, 'display.max_colwidth', 110):
    print(tabulate(df.sample(100)[["type_clean", "type", "title"]]))
```
